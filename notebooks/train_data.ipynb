```python
import pandas as pd

data = pd.read_csv('/home/serao/violence_detection/violence-detection/log/rlvs_r2plus1d_of/version_0/metrics.csv')
validation = data[['epoch','val_loss', 'val_acc']].dropna()
validation = validation.astype({'epoch':int})
validation.set_index('epoch', inplace=True)
train = data[['epoch','train_loss', 'train_acc']].dropna()
train = train.astype({'epoch':int})
train.set_index('epoch', inplace=True)
log = train.join(validation, on='epoch')
log
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,6))

log[['train_acc', 'val_acc']].plot(ax=axes[0])
axes[0].set_title('accuracy')
log[['train_loss', 'val_loss']].plot(ax=axes[1])
axes[1].set_title('loss')
```
